## Filters & Slicers

```python
df.columns
```

```python
# Top Categories by Profit
top_cat_profit = df.groupby(['Category'])['Profit'].sum().nlargest(3).reset_index()
top_cat_profit
```

```python
# What sub-category gained the most profit?
top_sub_profit = df[df['Category'] == "Technology"].groupby('Sub-Category')['Profit'].sum().nlargest(5).reset_index()
top_sub_profit
```

```python
# Top Categories according to Sales
top_categories_sales = df.groupby(['Category'])['Sales'].sum().nlargest(3).reset_index()
top_categories_sales
```

```python
# What sub-category had the highest sales?
top_sub_sales = df[df['Category'] == "Technology"].groupby('Sub-Category')['Sales'].sum(5).nlargest().reset_index()
top_sub_sales
```

```python
# State with the most profit
top_state_profit = df.groupby(['State'])['Profit'].sum().nlargest(10).reset_index()
top_state_profit
```

```python
# What State made the most sales?
top_state_sales = df.groupby(['State'])['Sales'].sum().nlargest(10).reset_index()
top_state_sales
```

```python
# In the state with the most sales, what city was the highest?
top_city_sales = df[df['State'] == "California"].groupby('City')['Sales'].sum().nlargest(5).reset_index()
top_city_sales
```

```python
# Segment with the highest average discount
top_seg_discn = df.groupby("Segment")["Discount"].mean().sort_values(ascending=False).reset_index()
top_seg_discn
```

```python
# What segment made the most profit?
top_seg_profit = df.groupby(['Segment'])['Profit'].sum().nlargest(3).reset_index()
top_seg_profit
```

```python
# What Customer accrued the most profit
top_10_customers = df.groupby(['Customer ID', 'Customer Name', 'Segment'])['Profit'].sum().nlargest(10).reset_index()
top_10_customers
```

## Save Filtered Results

```python
# 1
top_cat_profit.to_csv("top_category_by_profit.csv", index=False)
# 2
top_sub_profit.to_csv("subcategory_by_profit.csv", index=False)
# 3
top_categories_sales.to_csv("top_category_by_sales.csv", index=False)
# 4
top_sub_sales.to_csv("subcategory_by_sales.csv", index=False)
# 5
top_state_profit.to_csv("top_state_by_profit.csv", index=False)
# 6
top_state_sales.to_csv("top_state_by_sales.csv", index=False)
# 7
top_city_sales.to_csv("top_city_by_sales.csv", index=False)
# 8
top_seg_discn.to_csv("top_segment_by_discount.csv", index=False)
# 9
top_seg_profit.to_csv("top_segment_by_profit.csv", index=False)
# 10
top_10_customers.to_csv("top_10_customers_profit.csv", index=False)
```
